```python
import os

# Define the file path using os.path.join
file_path = os.path.join(path, "synthetic_stock_data.csv")

# Load the CSV file using NumPy
text_data = np.loadtxt(file_path, delimiter=',', skiprows=1, usecols=list(range(0,3)), dtype=str)
numeric_data = np.loadtxt(file_path, delimiter=',', skiprows=1, usecols=list(range(3,13)), dtype=float)
```
